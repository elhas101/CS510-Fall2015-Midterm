<h2>Discussion and observations</h2>

<h4>1. How does the plotted solution depend on your choice of time step size, and your choice of increment?  Do you see an improvement in precision from using the higher-order integration methods?</h4>

At step sizes below 200, second order runge-kutta method has the highest accuracy as seen below, whereas euler and runge kutta 4 fail to reach the same accuracy at step=80 for example.

<h3>rk2 and step=80</h3>

```python
RungeKutta = att.Attractor(end=80);
RungeKutta.evolve(order=2);
RungeKutta.plotx();
RungeKutta.plotxy();
RungeKutta.plotyz();
RungeKutta.plotzx();
RungeKutta.plot3d();
```

<h3>rk4 and step=80</h3>

```python
RungeKutta = att.Attractor(end=80);
RungeKutta.evolve(order=4);
RungeKutta.plotx();
RungeKutta.plotxy();
RungeKutta.plotyz();
RungeKutta.plotzx();
RungeKutta.plot3d();
```

At step above 300, we find that 4th order runge kutta is more accurate than rk2 and euler methods. However this accuracy receeds as step size is increased over 2000,

<h3>rk4 and step=1000</h3>

```python
RungeKutta = att.Attractor(end=1000);
RungeKutta.evolve(order=4);
RungeKutta.plotx();
RungeKutta.plotxy();
RungeKutta.plotyz();
RungeKutta.plotzx();
RungeKutta.plot3d();
```

<h3>rk2 and step=1000</h3>

```python
RungeKutta = att.Attractor(end=1000);
RungeKutta.evolve(order=2);
RungeKutta.plotx();
RungeKutta.plotxy();
RungeKutta.plotyz();
RungeKutta.plotzx();
RungeKutta.plot3d();
```

Euler method fails to return any accurate graphs for any range. 

In conclusion, higher order runge kutta functions don't always mean higher accuracy. While certainly any order increase over Euler method is an improvement, the accuracy returned by higher orders depend on the data analyzed and the step size. The 2nd order runge kutta may be better suited to a certain range in step size where runge kutta might not be, and vice versa for a higher step size.
